```python
import numpy as np
import matplotlib.pyplot as plt
import holoviews as hv
from holoviews import opts
from holoviews.streams import Selection1D
hv.extension('bokeh')
```

```python


alg = 'tsne'

wav = np.loadtxt('wav.txt',int)

path = "/Users/benjamin/Documents/postdoc/hetdex/hdr2.1.2/"

tsne_dat = np.load(path+alg+'/'+alg+'_components_hdr2.1.2.npy')
filenames,transformed = tsne_dat[:,0].astype('int').astype('str'),tsne_dat[:,1:]

cns_dat = np.load(path+'hdr2.1.2_contspec.npy')

spectra = cns_dat[:,1:].astype('float')
```

```python
#output_file('tsne_output.html', title='tSNE output')

tools = ['box_zoom', 'tap', 'reset']

```

```python
transformed_i = np.column_stack((transformed ,range(len(transformed))))
transformed_i = [tuple(transformed_i[j]) for j in range(len(transformed_i))]


spectral_data = hv.NdOverlay({int(i) : hv.Curve((wav,spectra[int(i)])) for _,_,i in transformed_i})
```

```python
points = hv.Points(transformed_i, [r'First '+alg+' feature',r'Second '+alg+' feature'])
```

```python
stream = Selection1D(source=points)
```

```python
empty = (hv.Points(np.random.rand(0,2)) * hv.Curve(np.random.rand(0,2))).relabel('No selection')


def get_spectra(index):
  
    if not index:
        print('No index!')
        print(' ')
        return empty
    else: print('There is an index!')
    spectrum = spectral_data[index[0]]
    curve =  hv.Curve(spectrum).relabel(filenames[index[0]])
    curve = curve.redim.label(x = r'Wavelength (\AA)', y = r'Flux (erg/s/cm^2)')
    return curve

selected_spectra = hv.DynamicMap(get_spectra, kdims=[], streams=[stream])

#average = hv.Curve(points).aggregate(function=np.mean)


#points.opts(tools=tools) + selected_spectra.opts(width=700)

layout = (selected_spectra + points).cols(1)

layout.opts(
	opts.Curve(color='royalblue', width=583, height=233, framewise=True, align='center',tools=['zoom_out']),
	#opts.Points(color='black', tools=tools, width=583,
    #            height=523,
	#	     marker='o', size=4, framewise=True, align='center', alpha=0.3),
    opts.Points(color='black',tools=tools, width=583,
                height=523,
		     marker='o', size=3, framewise=True, align='center', alpha=0.3),
	opts.Overlay(toolbar='above', legend_position='right',width=700))

#show(layout)

#fig = figure(title = 'tSNE', plot_height = 700, plot_width = 800,
#	     toolbar_location='below',
#	     tools=tools)

#fig.scatter(x = transformed[:,0], y = transformed[:,1],
#	   color='green', size=1.5, alpha=0.5)
  
#show(fig)
```
